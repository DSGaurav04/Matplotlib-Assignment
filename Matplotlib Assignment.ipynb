# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of  Matplotlib.

```python
#Line Plot:

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
#Scatter Plot:

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 5]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
#Histogram:

import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000) 

plt.hist(data, bins=30)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
#Pie Chart:

import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 15, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y, label='Data Points', color='green', marker='v')

plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()  
plt.grid(True)  
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.




```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])

y2 = np.array([50, 20, 40, 20, 60, 70])

y3 = np.array([10, 20, 30, 40, 50, 60])

y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2, 2, 1)  
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')

plt.subplot(2, 2, 2) 
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')

plt.subplot(2, 2, 3) 
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')

plt.subplot(2, 2, 4) 
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')

plt.tight_layout()

plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.




```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Bar Plot: Profit of Companies')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightcoral')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot: Profit of Companies')
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot: Comparison of Two Datasets')
plt.show()
```
